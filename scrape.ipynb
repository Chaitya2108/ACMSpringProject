```python
from ntscraper import Nitter
scraper = Nitter(log_level=1,skip_instance_check=False)
```

```python
terms = []
tweets = scraper.get_tweets("JoeBiden",mode="user",number=100)
```

```python
tweets["tweets"]
```

```python
import csv
f = open("biden_tweets.csv",'w')
writer = csv.writer(f)
```

```python
header = ['user','link','text','likes']
writer.writerow(header)

for tweet in tweets["tweets"]:
    data = [tweet['user']['name'],tweet['link'],tweet['text'],tweet['stats']['likes']]
    writer.writerow(data)
```
